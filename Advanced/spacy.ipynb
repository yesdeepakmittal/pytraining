```python
import spacy
nlp = spacy.load('en_core_web_sm')

def lemma(var):
    for token in var:
        print(f'{token.text:{12}} {token.pos_:{6}} {token.lemma:<{22}} {token.lemma_}')

inp_text = nlp(u"My computer has a good GPU power even then I use Azure cloud to train Deep Learning models.")

lemma(inp_text)
```
